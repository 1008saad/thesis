```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data: replace these with your actual model names, accuracies, and losses
model_names = ['Xception', 'VGG16', 'ResNet50', 'MobileNetV2', 'InceptionV3', 'InceptionResNetV2',
               'DenseNet201', 'DenseNet121', 'Ensemble-1', 'Ensemble-2']
accuracies = [0.9700, 0.9658, 0.9709, 0.9595, 0.9679, 0.9659, 0.9767, 0.9720, 0.9694, 0.9551 ]
losses = [0.1078, 0.1271, 0.1043, 0.1311, 0.1163, 0.1247, 0.0835, 0.0962, 0.1478, 1.2250]

# Convert accuracies to percentage
accuracies_percent = [acc * 100 for acc in accuracies]

# Create DataFrames for easier plotting with Seaborn
accuracy_data = {
    'Model': model_names,
    'Accuracy (%)': accuracies_percent
}
loss_data = {
    'Model': model_names,
    'Loss': losses
}
df_accuracy = pd.DataFrame(accuracy_data)
df_loss = pd.DataFrame(loss_data)

# Set the Seaborn style
sns.set(style="whitegrid")

# Create a bar chart for accuracies with a modern style
plt.figure(figsize=(12, 8))
sns.barplot(x='Model', y='Accuracy (%)', data=df_accuracy, palette="viridis")
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracies Comparison')
plt.ylim(0, 110)  # set y-axis limit to percentage scale
for i in range(len(model_names)):
    plt.text(i, accuracies_percent[i] + 2, f'{accuracies_percent[i]:.2f}%', ha='center')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Create a bar chart for losses with a modern style
plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Loss', data=df_loss, palette="viridis")
plt.xlabel('Models')
plt.ylabel('Loss')
plt.title('Model Losses Comparison')
plt.ylim(0, max(losses) + 0.5)  # add some space on top
for i in range(len(model_names)):
    plt.text(i, losses[i] + 0.02, f'{losses[i]:.2f}', ha='center')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
